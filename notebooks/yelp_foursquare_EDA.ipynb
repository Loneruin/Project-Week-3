```python
def restaurantPOI(lat,long):
    list_of_dict=[]   
    for x in restaurants['results']:
        categories = x.get('categories', None)
        cat_id = categories[0]['id'] if len(categories) > 0 else None
        category_name = categories[0]['name'] if len(categories) > 0 else None
        
        poi_dict = {
            'lat_long':f'{lat},{long}',
            'fsq_id':x['fsq_id'],
            'cat_id': cat_id,
            'category_name': category_name,
            'categories': categories,
            'name': x.get('name', None),
            'distance': x.get('distance', None),
            'latitude': x.get('geocodes',{}).get('main',{}).get('latitude',None),
            'longtitude': x.get('geocodes',{}).get('main',{}).get('longtitude',None),
            'address': x.get('location',{}).get('formatted_address', None),
            'popularity': x.get('popularity', None),
            'open_now': x.get('hours',{}).get('open_now', None),
            'rating': x.get('rating', None),
            'rating_count': x.get('stats',{}).get('total_ratings', None),
            'my_timestamp': datetime.now()
        }
        list_of_dict.append(poi_dict)
    return pd.DataFrame(list_of_dict)    
```
